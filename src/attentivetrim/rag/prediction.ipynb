```python
import json
import numpy as np

file_top50_uae = '/Users/chunwei/research/llm-scheduling/profiling-Meta-Llama-3-8B-Instruct-length-uae-top50.json'
file_top50_sfr = '/Users/chunwei/research/llm-scheduling/profiling-Meta-Llama-3-8B-Instruct-length-sfr-top50.json'

prediction_file = '/Users/chunwei/research/llm-scheduling/alpaca_chunwei_version.json'

with open(file_top50_uae, 'r') as f:
    top50_uae = json.load(f)
    
    uae_map = {}
    for item in top50_uae['results']:
        print(item)
        uae_map[item['record_id']] = {}
        uae_map[item['record_id']]['uae_average_5'] = item['top_50']['average_5']
        uae_map[item['record_id']]['uae_average_10'] = item['top_50']['average_10']
        uae_map[item['record_id']]['uae_average_20'] = item['top_50']['average_20']
        uae_map[item['record_id']]['uae_average_30'] = item['top_50']['average_30']
        uae_map[item['record_id']]['uae_average_40'] = item['top_50']['average_40']
        uae_map[item['record_id']]['uae_average_50'] = item['top_50']['average_length']
        
with open(file_top50_sfr, 'r') as f:
    top50_sfr = json.load(f)
    
    sfr_map = {}
    for item in top50_sfr['results']:
        print(item)
        sfr_map[item['record_id']] = {}
        sfr_map[item['record_id']]['sfr_average_5'] = item['top_50']['average_5']
        sfr_map[item['record_id']]['sfr_average_10'] = item['top_50']['average_10']
        sfr_map[item['record_id']]['sfr_average_20'] = item['top_50']['average_20']
        sfr_map[item['record_id']]['sfr_average_30'] = item['top_50']['average_30']
        sfr_map[item['record_id']]['sfr_average_40'] = item['top_50']['average_40']
        sfr_map[item['record_id']]['sfr_average_50'] = item['top_50']['average_length']
        
with open(prediction_file, 'r') as f:
    prediction = json.load(f)
    
    for item in prediction:
        id = item['id']
        if id in uae_map:
            item['uae_average_5'] = uae_map[id]['uae_average_5']
            item['uae_average_10'] = uae_map[id]['uae_average_10']
            item['uae_average_20'] = uae_map[id]['uae_average_20']
            item['uae_average_30'] = uae_map[id]['uae_average_30']
            item['uae_average_40'] = uae_map[id]['uae_average_40']
            item['uae_average_50'] = uae_map[id]['uae_average_50']
        if id in sfr_map:
            item['sfr_average_5'] = sfr_map[id]['sfr_average_5']
            item['sfr_average_10'] = sfr_map[id]['sfr_average_10']
            item['sfr_average_20'] = sfr_map[id]['sfr_average_20']
            item['sfr_average_30'] = sfr_map[id]['sfr_average_30']
            item['sfr_average_40'] = sfr_map[id]['sfr_average_40']
            item['sfr_average_50'] = sfr_map[id]['sfr_average_50']
    
    with open('/Users/chunwei/research/llm-scheduling/alpaca_chunwei_version_with_similarity.json', 'w') as f:
        json.dump(prediction, f, indent=4)
```
